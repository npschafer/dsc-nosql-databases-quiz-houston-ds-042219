## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

SQL databases are relational databases, while NoSQL databases are not relational databases. Whether SQL or NoSQL databases are better will depend on the data being stored and how the data will be accessed.





## Question 2

NoSQL databases are preferrable to SQL databases when, for example, the data being stored does not fit well into table format. An example of data that would not fit well into tables would be long documents of variable length.




## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MongoDB is a document store database where the documents are represented as sets of key-value pairs, with the data being stored in the values. Unlike a relational database where data is accessed by table, row, and column, to access data in MongoDB or other document store-type databases, you need to know what key to look for, which is information that would usually be provided in a schema for the database.







## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

When data gets large, it becomes necessary and also sometimes advantageous to store data across multiple computers (or "servers"). Calculations that are performed on data sets that are distributed across multiple computers need to, in the end, be combined to compute a final result (where this final combination of results is often referred to as "reduce" operation). If one or more of the computers was offline, the final reduced computation would be incorrect. Fault redudancy ensures that the entire data set will be used in the final computation by storing backups of parts of the database on other computers on the network.
